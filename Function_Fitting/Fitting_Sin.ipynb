```python
def read_data(filename):
    '''
    读取数据
    '''
    with open(filename,'r') as f:
        line=list(map(float,f.read().replace('\n','').replace('{','').replace('}','').split(',')))
    data=[]
    for i in range(len(line)//2):
        data.append((line[2*i],line[2*i+1]))
    return pd.DataFrame(data)
```

```python
def obj_func(com,C):
    x=com[0]
    A=com[1]
    B=com[2]
    return A*np.cos(B*x+C)
```

```python
def find_hills(arr,reverse=False):
    '''
    寻找函数峰值
    返回峰值数值和bool声明序列
    '''
    arr=np.array(arr)
    is_hill=np.zeros(len(arr))
    for i in range(1,len(arr)-1):
        if reverse:
            if arr[i-1]>arr[i]<arr[i+1]:
                is_hill[i]=1
        else:
            if arr[i-1]<arr[i]>arr[i+1]:
                is_hill[i]=1
    return (arr[is_hill==1],is_hill)
```

```python
def find_period(time_arr,mask):
    indexes=np.arange(len(mask))[mask==1]
    periods=[]
    for i in range(len(indexes)-1):
        periods.append(time_arr[indexes[i+1]]-time_arr[indexes[i]])
    return (periods,sum(periods)/len(periods))
```

```python
def sync(a,A,B,th,e,mu):
    '''
    同相位函数
    '''
    result=-2*a+2*np.sin(a)-e*(np.sqrt(abs(A**2-a**2))-np.sqrt(abs(B**2-a**2)))*(1-a**2/th**2)\
-2*mu*np.cos(a)*((2*a**2-A**2-B**2)*np.sin(a)-2*a*np.cos(a))
    return result
```

```python
def asyc(b,A,B,th,e,mu,delta=0.23,r=0.05):
    '''
    反相位函数
    '''
    result=-2*b+2*np.sin(b)-e*(np.sqrt(abs(A**2-b**2))+np.sqrt(abs(B**2-b**2)))*(1-b**2/th**2)\
   # -2*r*np.sin(b)+0.23
    return float(result)
```

```python
def visualize(left_data,right_data,slic,shape=(20,8),title=''):
    f=plt.figure(dpi=120,figsize=shape)
    f.dpi=120
    if slic is None:
        start,end=0,len(left_data)
    else:
        start,end=slic
    plt.plot(left_data[0][start:end],left_data[1][start:end],marker='o',color='c')
    plt.plot(right_data[0][start:end],right_data[1][start:end],marker='o',color='orange')
    plt.xlabel('Time/s')
    plt.ylabel('Angle/rad')
    plt.title('Amplitude-Time Relationships'+title)
    #可视化
```

```python
def wraper(side):
    time_data=read_data(side+'_time_data.txt')
    angle_data=read_data(side+'_angle_data.txt')
    data=pd.DataFrame(list(zip(time_data[1],angle_data[1])))
    return data
```

```python
def reader(num):
    time_data=read_data('time.txt')
    angle_data=read_data('2170({})angle.txt'.format(num))
    data=pd.DataFrame(list(zip(time_data[1],angle_data[1])))
    return data
```

```python
def main(estimate,left_data,right_data,slic,mu=1,which=0,drop_edge=False):
    start,end=slic
    d1=left_data[start:end][1]
    d2=right_data[start:end][1]
    A=find_hills(d1)[0].mean()
    B=find_hills(d2)[0].mean()
    a=np.array((d1+d2)/2)
    b=np.array((d1-d2)/2)
    if drop_edge:
        a=a[np.abs(a)<0.5]
        b=b[np.abs(b)<0.5]
    e=0.2
    th=1.
    alpha=(a,b)[which]
    return pd.Series(list(map(lambda x: estimate(x,A,B,th,e,mu),alpha)))
```

```python
def error_estimate(data1,data2,mu,sync=(300,390),asyc=(932,992)):
    '''
    data1,data2: the left and right side data
    mu: the value of mu
    sync: the slice of sync range
    asyc: the slice of asyc range
    
    eastimate the error of the two data
    '''
    arr=main(sync,data1,data2,sync,mu,which=0)
    print(arr.dropna().describe())
    arr.plot.density()
    plt.show()
    arr=main(asyc,data1,data2,asyc,mu,which=1)
    print(arr.dropna().describe())
    arr.plot.density()
    plt.show()
```
